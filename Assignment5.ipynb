# splitting into data sets

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df_filled.iloc[:,0:2],df_filled.iloc[:,1],test_size=0.2)
x_train1,x_test1,y_train1,y_test1=train_test_split(df_filled.iloc[:,3:],df_filled.iloc[:,1],test_size=0.2)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train1)
x_train_scaled=scaler.transform(x_train1)
x_test_scaled=scaler.transform(x_test1)
```
